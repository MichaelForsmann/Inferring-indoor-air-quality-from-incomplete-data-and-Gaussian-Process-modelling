```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
smoke=pd.read_excel(r"\\192.168.1.59\Data\COPSAC-2000\Nikotin-kotinin\Nikotin-kotinin 27-05-2005 plots.xls")
```

```python
smoke[smoke=="md"]=0
```

```python
pm25=pd.read_excel(r"pm25_BC_corrected.xlsx")
```

```python
pm25_1=pm25.loc[pm25.maalnr==1]
pm25_2=pm25.loc[pm25.maalnr==2]
pm25_3=pm25.loc[pm25.maalnr==3]
```

```python
pm25
```

```python
smoke_1=smoke.set_index("Copsac-nr").loc[:,["nikotin plasma (ng/ml), 6 mdr","kotinin plasma (ng/ml), 6 mdr"]]
smoke_2=smoke.set_index("Copsac-nr").loc[:,["nikotin hår (ng/mg), 1 år","kotinin hår (ng/mg), 1 år"]]
```

```python
pm25_1=pm25_1.set_index("bnr")
pm25_1.index.name="Copsac-nr"
pm25_2=pm25_2.set_index("bnr")
pm25_2.index.name="Copsac-nr"
pm25_3=pm25_3.set_index("bnr")
pm25_3.index.name="Copsac-nr"
```

```python
data_1=pd.concat([smoke_1,pm25_1],axis=1)
data_2=pd.concat([smoke_2,pm25_2],axis=1)
data_3=pd.concat([smoke_2,pm25_2],axis=1)
```

```python
data_1=data_1.rename(columns={"nikotin plasma (ng/ml), 6 mdr":"nikotin","kotinin plasma (ng/ml), 6 mdr":"kotinin"})
data_2=data_2.rename(columns={"nikotin hår (ng/mg), 1 år":"nikotin","kotinin hår (ng/mg), 1 år":"kotinin"})
data_3=data_3.rename(columns={"nikotin hår (ng/mg), 1 år":"nikotin","kotinin hår (ng/mg), 1 år":"kotinin"})
```

```python
data_3.columns
```

```python
data_all=pd.concat([data_1,data_2,data_3])
```

```python
list2=["sv_final","pm_final",'motorway','road','powerplant','Area','flow',"nikotin","kotinin",'cookerhood','passive']
```

```python
spearman=data_all.corr(method="spearman").loc[list2,["pm25_Gaussion","pm25_linear","pm25_exp","pm25_log",'BC_Gaussion', 'BC_linear', 'BC_exp', 'BC_log']]
```

```python
pearson=data_all.corr().loc[list2,["pm25_Gaussion","pm25_linear","pm25_exp","pm25_log",'BC_Gaussion', 'BC_linear', 'BC_exp', 'BC_log']]             
```

```python
spearman.index=spearman.index+" spearman"
```

```python
corr=pd.concat([pearson,spearman]).sort_index()
```

```python
corr.index=corr.index.str.upper()
corr=corr.rename(index={"PM_FINAL":"Raw_PM25","PM_FINAL SPEARMAN":"Raw_PM25 SPEARMAN","SV_FINAL":"Raw_BC","SV_FINAL SPEARMAN":"Raw_BC SPEARMAN"})
```

```python
fig=sns.heatmap(corr,annot=True)
fig=fig.get_figure()
fig.savefig("gaussion_correction_correlation.png")
```
