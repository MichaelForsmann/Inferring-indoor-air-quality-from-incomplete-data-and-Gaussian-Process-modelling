```python
data=pd.read_excel(r"../data/pm25_BC_corrected.xlsx")
```

```python
# Extract week,dat
no2["sadto"]=pd.to_datetime(no2.start_n)
no2=no2.assign(week=no2.sadto.apply(lambda x: x.week))
no2=no2.assign(month=no2.sadto.apply(lambda x: x.month))
no2=no2.assign(year=no2.sadto.apply(lambda x: x.year))
no2=no2.assign(day=no2.sadto.apply(lambda x: x.day_of_year))
```

```python
data_bnr=data.set_index("bnr")
data_bnr=data_bnr.drop(['flow', 'pm_final', 'sv_final','week', 'month', 'year', 'day',
       'corrected_week', 'label', 'BC_Gaussion', 'BC_linear', 'BC_exp',
       'BC_log', 'pm25_Gaussion', 'pm25_linear', 'pm25_exp', 'pm25_log'],axis=1)
data_bnr.index.name="cpnr"
data_bnr_1=data_bnr.loc[data_bnr.maalnr==1,:]
data_bnr_2=data_bnr.loc[data_bnr.maalnr==2,:]
data_bnr_3=data_bnr.loc[data_bnr.maalnr==3,:]
```

```python
#removeig unwanted stuff

no2=no2.replace(",",".",regex=True).rename(columns={"no2_(μg/m3)":"no2","no_(μg/m3)":"Nox","formald_(μg/m3)":"formaldehyd","acetald_(μg/m3)":"acetald","acetone_(μg/m3)":"acetone"})
no2.no2=no2.no2.astype(float)
no2.Nox=no2.Nox.astype(float)
```

```python
no2.loc[no2.visit=="6mdr","visit"]=1
no2.loc[no2.visit=="1yr","visit"]=2
no2.loc[no2.visit=="18mdr","visit"]=3
no2.loc[no2.visit=="2yr","visit"]=4
no2.loc[no2.visit=="30mdr","visit"]=5
no2.loc[no2.visit==18,"visit"]=5
no2.loc[no2.visit=="?","visit"]=4
```

```python
no2["corrected_week"]=no2.day/7
```

```python
no2=no2.drop(['sadto', 'week', 'month','batch_c', 'start_c', 'stop_c',
       'batch_n', 'start_n', 'stop_n','Unnamed: 0','year','day','Nox'],axis=1)
```

```python
no2_cpnr=no2.set_index("cpnr")
no2_cpnr_1=no2_cpnr.loc[no2_cpnr.visit==1]
no2_cpnr_2=no2_cpnr.loc[no2_cpnr.visit==2]
no2_cpnr_3=no2_cpnr.loc[no2_cpnr.visit==3]
no2_cpnr_4=no2_cpnr.loc[no2_cpnr.visit==4]
no2_cpnr_5=no2_cpnr.loc[no2_cpnr.visit==5]
visit=pd.concat([data_bnr_1,no2_cpnr_1],axis=1).reset_index()
visit2=pd.concat([data_bnr_2,no2_cpnr_2],axis=1).reset_index()
visit3=pd.concat([data_bnr_3,no2_cpnr_3],axis=1).reset_index()
full_data_no2=pd.concat([visit,visit2,visit3]).dropna(subset="no2_corrected")
```

```python
imputer = IterativeImputer(random_state=0)
no2_1=imputer.fit_transform(scaler.fit_transform(full_data_no2))
no2_2=scaler.inverse_transform(no2_1)
no2_df=pd.DataFrame(np.round(no2_2,1),columns=full_data_no2.columns,index=full_data_no2.cpnr)
```

```python
no2_df=no2_df.where(no2_df>0,0)
```

```python
no2_df.to_excel("../data/before_season_correction.xlsx")
```

```python
cpnr=no2_df.loc[:,["sample_time","corrected_week","gastove","passive","road","sample_time","corrected_week"]].groupby("cpnr").mean().reset_index()
X_train, X_test, y_train, y_test = ks.train_test_split(cpnr.loc[:,["gastove","passive","road","sample_time","corrected_week"]], cpnr.cpnr, test_size = 0.2);
no2_df=no2_df.set_index("cpnr")
train=no2_df.loc[y_train.values,:].reset_index()
test=no2_df.loc[y_test.values,:].reset_index()
```

```python
extra_data_end=train.loc[train.corrected_week<20,:]
extra_data_start=train.loc[train.corrected_week>52.3-20,:]
extra_data_end.corrected_week=extra_data_end.corrected_week+52.3
extra_data_start.corrected_week=-(52.3-extra_data_start.corrected_week)
extra_data_start=extra_data_start.set_index(np.zeros(extra_data_start.shape[0]))
extra_data_end=extra_data_end.set_index(np.zeros(extra_data_end.shape[0]))
```

```python
train=pd.concat([train,extra_data_end,extra_data_start])
train_pm=pd.concat([train_pm,extra_data_end_pm,extra_data_start_pm])
```
