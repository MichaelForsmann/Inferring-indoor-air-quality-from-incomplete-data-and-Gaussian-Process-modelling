```python
import numpy as np

from bokeh.plotting import figure, show
import pandas as pd
from bokeh.io import curdoc,output_notebook,export_png
from bokeh.layouts import column,gridplot
#from print_versions import print_versions
from sklearn.metrics import r2_score
from seaborn import clustermap
from bokeh.models import Band, ColumnDataSource
import pandas as pd
TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,examine,help"
```

```python
PM25=pd.read_excel("../data/Pre_flow_cal.xlsx")
```

```python
output_notebook()
```

```python
PM25_smoke=PM25.loc[PM25.passive>0]
PM25_nonsmoke=PM25.loc[PM25.passive==0]
```

```python
no2=pd.read_excel(r"../data/before_season_correction.xlsx")
```

```python
no2_smoke=no2.loc[PM25.passive>0]
no2_nonsmoke=no2.loc[PM25.passive==0]
```

```python
no2_smoke
```

```python


rng = np.random.default_rng()
x = rng.normal(loc=0, scale=1, size=1000)

p = figure(width=500, height=500, toolbar_location=None,
           title="Normal (Gaussian) Distribution")

# Histogram
bins = np.linspace(0, 120, 40)
hist_smoke, edges_smoke = np.histogram(PM25_smoke.pm_final, density=True, bins=bins)
hist, edges = np.histogram(PM25_nonsmoke.pm_final, density=True, bins=bins)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="skyblue", line_color="white",
         legend_label="nonsmoker",fill_alpha =0.7)
p.quad(top=hist_smoke, bottom=0, left=edges_smoke[:-1], right=edges_smoke[1:],
         fill_color="red", line_color="white",
         legend_label="smoker",fill_alpha =0.5)
p.y_range.start = 0
p.xaxis.axis_label = "$$PM_{2.5}$$"
p.yaxis.axis_label = "PDF(x)"
p1 = figure(width=500, height=500, toolbar_location=None,
           title="Normal (Gaussian) Distribution")

# Histogram
bins_bc = np.linspace(0, 4, 40)
hist_smoke_bc, edges_smoke_bc  = np.histogram(PM25_smoke.sv_final, density=True, bins=bins_bc)
hist_bc , edges_bc  = np.histogram(PM25_smoke.sv_final, density=True, bins=bins_bc)

p1.quad(top=hist_bc , bottom=0, left=edges_bc [:-1], right=edges_bc [1:],
         fill_color="skyblue", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.7)
p1.quad(top=hist_smoke_bc, bottom=0, left=edges_smoke_bc [:-1], right=edges_smoke_bc [1:],
         fill_color="red", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.5)
p1.y_range.start = 0
p1.xaxis.axis_label = "Black carbon"
p1.yaxis.axis_label = "PDF(x)"
p2 = figure(width=500, height=500, toolbar_location=None,
           title="Normal (Gaussian) Distribution")

# Histogram
bins_no2 = np.linspace(0, 40, 40)
hist_smoke_no2 , edges_smoke_no2  = np.histogram(no2_smoke.no2, density=True, bins=bins_no2)
hist_no2 , edges_no2  = np.histogram(no2_nonsmoke.no2, density=True, bins=bins_no2)

p2.quad(top=hist_no2 , bottom=0, left=edges_no2 [:-1], right=edges_no2 [1:],
         fill_color="skyblue", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.7)
p2.quad(top=hist_smoke_no2 , bottom=0, left=edges_smoke_no2 [:-1], right=edges_smoke_no2 [1:],
         fill_color="red", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.5)
p2.y_range.start = 0
p2.xaxis.axis_label = "x"
p2.yaxis.axis_label = "PDF(x)"
p3 = figure(width=500, height=500, toolbar_location=None,
           title="Normal (Gaussian) Distribution")

# Histogram
bins_nox = np.linspace(0, 120, 40)
hist_smoke_nox , edges_smoke_nox  = np.histogram(no2_smoke.nox, density=True, bins=bins_nox)
hist_nox , edges_nox  = np.histogram(no2_nonsmoke.nox, density=True, bins=bins_nox)

p3.quad(top=hist_nox , bottom=0, left=edges_nox [:-1], right=edges_nox[1:],
         fill_color="skyblue", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.7)
p3.quad(top=hist_smoke_nox , bottom=0, left=edges_smoke_nox [:-1], right=edges_smoke_nox [1:],
         fill_color="red", line_color="white",
         legend_label="1000 random samples",fill_alpha =0.5)
p3.y_range.start = 0
p3.xaxis.axis_label = "$$NO_X$$"
p3.yaxis.axis_label = "PDF(x)"
grid =  gridplot([[p, p1],[p2, p3]], width=500, height=500);

show(column(grid));
```
