```python
import torch
import arviz as az
from scipy.optimize import curve_fit
from bokeh.plotting import figure, show,output_file, save
from bokeh.transform import factor_cmap, factor_mark
from bokeh.palettes import Spectral
from bokeh.models import Slope, Div
from bokeh.io import curdoc,output_notebook,export_png
from bokeh.layouts import column,gridplot
import numpy as np
from sklearn.metrics import r2_score
#from seaborn import clustermap*
from bokeh.models import Band, ColumnDataSource
import pandas as pd
TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,examine,help"
```

```python
PM25=pd.read_excel("../data/Pre_flow_cal.xlsx")
train,test=PM25.loc[PM25.label==0,:],PM25.loc[PM25.label==1,:]
```

```python
gpr=torch.load("../models/PM2.5_flow",weights_only=False,map_location='cuda')
gpr_bc=torch.load("../models/bc_flow",weights_only=False,map_location='cuda')
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
linmod=torch.linspace(3.6,4.4,800).to(device)
pm25_model,pm25_model_std=gpr(linmod,full_cov=True)

linmod_np=linmod.cpu().detach().numpy()
pm25_model_np,pm25_model_std_np=pm25_model.cpu() .detach().numpy().copy(),pm25_model_std.diag().sqrt().cpu() .detach().numpy().copy()

lower1 = pm25_model_np - pm25_model_std_np
upper1 = pm25_model_np + pm25_model_std_np
data1=pd.DataFrame([linmod_np,lower1,upper1],index=["flow","lower","upper"]).T
data1 = ColumnDataSource(data1.reset_index())
```

```python

bc_model,bc_model_std=gpr_bc(linmod, full_cov=True)
bc_model_np,bc_model_std_np=bc_model.cpu().detach().numpy().copy(),bc_model_std.cpu().diag().sqrt().detach().numpy().copy()

lower1_bc = bc_model_np - bc_model_std_np
upper1_bc = bc_model_np + bc_model_std_np
data1_bc=pd.DataFrame([linmod_np,lower1_bc,upper1_bc],index=["flow","lower_bc","upper_bc"]).T
data1_bc = ColumnDataSource(data1_bc.reset_index())
```

```python
def roll_week(data,resolution,week,particle,parameter,x):
    mean=[]
    std=[]
    sort=data.sort_values(parameter)
    list_std=[particle+"_std" for particle in particle]
    for i in x:
        mean.append(sort.loc[(i-week<sort.flow)&(i+week>sort.flow),particle].mean())
        std.append(sort.loc[(i-week<sort.flow)&(i+week>sort.flow),particle].std())
    mean_1,std_1=pd.DataFrame(mean,index=x),pd.DataFrame(std,index=x)
    std_1.columns=std_1.columns+"_std"
    data=pd.concat([mean_1,std_1],axis=1)
    return data.set_index(x)
```

```python
resolution=500
x=np.linspace(3.7,4.3,resolution,endpoint=True)
mean=roll_week(train,500,0.2,["sv_final","pm_final"],"flow",x)
mean_test=roll_week(test,500,0.2,["sv_final","pm_final"],"flow",x)
```

```python

lower1_std = mean.pm_final - mean.pm_final_std
upper1_std = mean.pm_final + mean.pm_final_std
lower1_bc_std  = mean.sv_final - mean.sv_final_std
upper1_bc_std  = mean.sv_final + mean.sv_final_std
mean["lower1_std_pm"]=lower1_std
mean["upper1_std_pm"]=upper1_std
mean["lower1_std_bc"]=lower1_bc_std
mean["upper1_std_bc"]=upper1_bc_std

mean1 = ColumnDataSource(mean.reset_index())
```

```python
train_bc=gpr_bc(torch.tensor(mean.index.values).float().to(device))[0].cpu().detach().numpy()
train_pm=gpr(torch.tensor(mean.index.values).float().to(device))[0].cpu().detach().numpy()
```

```python
r2_bc,r2_pm=r2_score(mean.sv_final,train_bc),r2_score(mean.pm_final,train_pm)
r2_bc_test,r2_pm_test=r2_score(mean_test.sv_final,train_bc),r2_score(mean_test.pm_final,train_pm)
```

```python
r2_bc,r2_pm,r2_bc_test,r2_pm_test
```

```python
p = figure(x_range=(3.6,4.3),y_range=(0,60));
p.title.text = r"$$ PM_{2.5}$$ calibration of flow";
p.xgrid.grid_line_color=None;
p.ygrid.grid_line_alpha=0.5;

p.scatter(train.flow, y=train.pm_final, color="blue", marker="dot", size=20, alpha=0.4,legend_label="raw points");
band = Band(base="index", lower="lower1_std_pm", upper="upper1_std_pm",source=mean1, fill_color="red", line_color="black",fill_alpha=0.2);
band1 = Band(base="flow", lower="lower", upper="upper",source=data1,fill_alpha=0.5, fill_color="blue", line_color="black");

p.add_layout(band);
p.add_layout(band1);
p.line(mean.index, mean.pm_final, line_width=3,color="green",legend_label="Train roll mean R2 "+str(round(r2_pm,2)));
p.line(mean_test.index, mean_test.pm_final, line_width=3,color="orange",legend_label="test roll mean R2 "+str(round(r2_pm_test,2)));

p.line(linmod_np, pm25_model_np, line_width=3,color="red",legend_label="Gaussion model");
p.xaxis.axis_label = r'$$\frac{L}{s}$$';
p.yaxis.axis_label = r'$$\frac{\mu g}{m^3} $$';
p.yaxis.axis_label_orientation  = 0
p1 = figure(x_range=(3.6,4.3),y_range=(0,3));
p1.title.text = r" Black Carbon calibration of flow";
p1.xgrid.grid_line_color=None;
p1.ygrid.grid_line_alpha=0.5;

p1.scatter(train.flow, y=train.sv_final, color="blue", marker="dot", size=20, alpha=0.8,legend_label="raw points");
band2 = Band(base="flow", lower="lower_bc", upper="upper_bc",source=data1_bc, fill_color="blue", line_color="black",fill_alpha=0.4);
band3 = Band(base="index", lower="lower1_std_bc", upper="upper1_std_bc",source=mean1, fill_color="red", line_color="black",fill_alpha=0.2);

p1.add_layout(band2);
p1.add_layout(band3);
p1.line(mean_test.index, mean_test.sv_final, line_width=3,color="orange",legend_label="test roll mean R2 "+str(round(r2_bc_test,2)));
p1.line(mean.index, mean.sv_final, line_width=3,color="green",legend_label="train roll mean R2 "+str(round(r2_bc,2)));
p1.line(linmod_np, bc_model_np, line_width=3,color="red",legend_label="Gaussion model");
p1.yaxis.axis_label_orientation  = 0

p1.xaxis.axis_label = r'$$\frac{L}{s}$$';
p1.yaxis.axis_label = r'$$ \frac{\mu g}{m^3} $$';
p1.legend.title_text_font_size = "16px";
p.legend.title_text_font_size = "16px";
p.xaxis.axis_label_text_font_size = "14px";
p.yaxis.axis_label_text_font_size = "14px";
p1.xaxis.axis_label_text_font_size = "14px";
p1.yaxis.axis_label_text_font_size = "14px";
p1.yaxis.axis_label_text_font_style = "bold";
p1.xaxis.axis_label_text_font_style = "bold";
p.yaxis.axis_label_text_font_style = "bold";
p.xaxis.axis_label_text_font_style = "bold";
grid =  gridplot([[p, p1]], width=500, height=500);

show(column(grid));
```
